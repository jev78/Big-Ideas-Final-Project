# Fire Incidents: Least Incidents - Joey Vojtas

To help us determine the best neighborhood in Pittsburgh, we looked at all of the fire incidents in the city.  We decided that we would rank the neighborhoods with the most fires higher than the ones that have less fires.  This data will then help us determine the best neighborhood overall in conjunction with the other data sets.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

fire_data = pd.read_csv("City_Of_Pitt._Fire_Incidents.csv")

filtered_data = fire_data[fire_data['neighborhood'].notna()]
incident_counts = filtered_data['neighborhood'].value_counts().reset_index()

incident_counts.columns = ['Neighborhood', 'Fire Incidents']

reversed_incident_counts = incident_counts.sort_values(by='Fire Incidents', ascending=False)

reversed_incident_counts.index += 1
reversed_incident_counts.index.name = "Rank"

print(reversed_incident_counts)
```

As we can see here there is tons of data on fire incidents around Pittsburgh, but at the top of the list is Carrick.  The Central Business District is also present near the top.  This is important because it placed in first in war memorials.

## Graph

Below is the data from the table in graph formatting.

```python
plt.figure(figsize=(12, 8))
plt.bar(reversed_incident_counts['Neighborhood'], reversed_incident_counts['Fire Incidents'], color='lightcoral', edgecolor='black')
plt.xlabel('Neighborhood', fontsize=12)
plt.ylabel('Number of Fire Incidents', fontsize=12)
plt.title('Fire Incidents by Neighborhood in Pittsburgh', fontsize=14)
plt.xticks(rotation=45, ha='right', fontsize=7) 
plt.yticks(fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

As you can see from the graph its data is the same as the table with the Carrick at the top with a few others and then a steady decline.
