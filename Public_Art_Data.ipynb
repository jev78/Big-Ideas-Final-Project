# Public Art: War Memorials - Matthew Caputo
To help us determine the best neighborhood in Pittburgh, we looked at all of the public art works and decided that the neighborhood with the most war memorials would be deemed the best neighborhood in this category.  This data will then help us determine the best neighborhood overall in conjunction with the other data sets.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

art_data = pd.read_csv("City_Of_Pitt._Public_Art_Data.csv") 
war_memorials = art_data[art_data['art_type'] == "War Memorials"]

war_memorials_count = war_memorials['neighborhood'].value_counts().reset_index()
war_memorials_count.columns = ['Neighborhood', 'War Memorials']

war_memorials_count_sorted = war_memorials_count.sort_values(by='War Memorials', ascending=False)
war_memorials_count_sorted = war_memorials_count_sorted.reset_index(drop=True)

war_memorials_count_sorted.index += 1
war_memorials_count_sorted.index.name = "Rank"

print(war_memorials_count_sorted)
```

As we can see here many neighborhoods have only one war memorial, but as we keep going up the table we can see that alone at the top with 5 war memorials is the Central Buisness District

## Graph
Below is the data from the table in graph formatting.

```python
plt.figure(figsize=(12, 6))
plt.bar(war_memorials_count_sorted['Neighborhood'], war_memorials_count_sorted['War Memorials'])
plt.title('Number of War Memorials by Neighborhood')
plt.xlabel('Neighborhood')
plt.ylabel('Number of War Memorials')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

As you can see from the graph its data is the same as the table with the Central Buisness District at the top and then a slow decline in war memorials.
