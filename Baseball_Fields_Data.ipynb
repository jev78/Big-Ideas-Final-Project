# Baseball Fields: Most Lights - Shared Notebook
To help us determine the best neighborhood in Pittburgh, we looked at all of the baseball fields present in the city.  We decided that we would rank the neighborhoods from the ones with the most fields with the most lights to the lowest amount of fields with lights.  This data will then help us determine the best neighborhood overall in conjunction with the other data sets.

To start below is the code to create a table that ranks the neighborhoods based off of the ones with the most fields.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

baseball_fields = pd.read_csv("City_Of_Pitt_Baseball_Fields.csv")

neighborhood_field_counts = baseball_fields['Neighborhood'].value_counts()

neighborhood_field_counts_df = neighborhood_field_counts.reset_index()
neighborhood_field_counts_df.columns = ['Neighborhood', 'Total Fields']

neighborhood_field_counts_df = neighborhood_field_counts_df.sort_values(by='Total Fields', ascending=False)

neighborhood_field_counts_df.insert(0, 'Rank', range(1, len(neighborhood_field_counts_df) + 1))
neighborhood_field_counts_df.reset_index(drop=True, inplace=True)

print(neighborhood_field_counts_df.to_string(index=False))
```

Building apon this, below is the new table that ranks the neighborhoods in order of fields with lights.

```python
fields_with_lights_one = baseball_fields[baseball_fields['Lights'] == 1]

neighborhood_lights_counts = fields_with_lights_one['Neighborhood'].value_counts()

neighborhood_lights_counts_df = neighborhood_lights_counts.reset_index()
neighborhood_lights_counts_df.columns = ['Neighborhood', 'Fields with Lights']

neighborhood_lights_counts_df = neighborhood_lights_counts_df.sort_values(by='Fields with Lights', ascending=False)

neighborhood_lights_counts_df.insert(0, 'Rank', range(1, len(neighborhood_lights_counts_df) + 1))
neighborhood_lights_counts_df.reset_index(drop=True, inplace=True)

print(neighborhood_lights_counts_df.to_string(index=False))
```

Based off of our new found data we see Mount Washington and Southside Slopes jump to the top.  We also see the former number 1 Squirrel hill South fall to 14th.

## Graph
Below is the data from thenew  table in graph formatting.

```python
plt.figure(figsize=(12, 6))
plt.bar(neighborhood_lights_counts_df['Neighborhood'], neighborhood_lights_counts_df['Fields with Lights'], color='skyblue')
plt.title('Number of Fields with Lights by Neighborhood', fontsize=14)
plt.xlabel('Neighborhood', fontsize=12)
plt.ylabel('Fields with Lights', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout() 
plt.show()
```

As you can see from the graph its data is the same as the new table with Mount Washington at the top with 4, and then a decline after that.
